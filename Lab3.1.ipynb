```python
import matplotlib.pyplot as plt
import numpy as np
def f(x):
    return x**2
def df(x):
    return 2*x
def g(x):
    return x**4 + x**3 + x**2
def dg(x):
    return 4*(x**3) + 3*(x**2) + 2*(x)
def cos(x):
    return np.cos(np.pi*x)
def dcos(x):
    return (-1 )* (np.pi)*np.sin(np.pi * x)

def batch_gradient_descent(initial_x, lr, iterations, gradient):
    x = initial_x
    trajectory = [x]
    for _ in range(iterations):
        grad = gradient(x)
        #print(grad)
        x -= lr * grad
        trajectory.append(x)
    #print(trajectory)
    return trajectory

def plotGradientDescent(initialValues,x ,fArray, dfArray, label, title, learningRates, epoch):
    fig, axs = plt.subplots(1, len(fArray),figsize = (6*len(fArray), 2*len(fArray) ))
    for item in range(len(fArray)):
        trajectory =  batch_gradient_descent(initialValues[item], learningRates[item], epoch, dfArray[item])
        axs[item].plot(x[item], fArray[item](x[item]), label= label[item])
        axs[item].scatter(trajectory, fArray[item](np.array(trajectory)), c='red', label=f"steps\n (Learning Rate : {learningRates[item]})")
        for i in range(len(trajectory) - 1):
            axs[item].annotate('', xy=(trajectory[i + 1], dfArray[item](trajectory[i + 1])),
                            xytext=(trajectory[i], dfArray[item](trajectory[i])))
        #axs[item].set_title(f"{title[item]} (Learning Rate: {learningRates[item]})")
        axs[item].set_title(title[item])
        axs[item].set_xlabel("x")
        axs[item].set_ylabel("Cost")
        axs[item].legend()

functionArray = [f,g,cos]
dfArray =[ df,dg, dcos]
label =["Cost Function: x^2", "Cost function :x^4+ x^3 + x^2 ", "Cost Function of cos(πx)"]
title = ["Batch Gradient Descent: x^2", "Batch Gradient Descent: x^4+ x^3 + x^2", " Batch Gradient Descent : cos(πx)"]
x = np.linspace(-5,5,100)
#print(x)
x1 = np.linspace(-5,5 ,100)
x2 = np.linspace(-2,2,100)
xvalues = [x, x1,x2]
intialValues = [-6, -4 , -0.5 ]
learningRates =[ 0.1, 0.001, 0.1]
epoch = 40
plotGradientDescent(intialValues,xvalues,functionArray,dfArray,label, title, learningRates,epoch)
```
