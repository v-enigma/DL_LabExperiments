```python
def f(x):
    return x**2
def df(x):
    return 2*x
def g(x):
    return x**4 + x**3 + x**2 

def dg(x):
    return 4*(x**3) + 3*(x**2) + 2*(x)

def batch_gradient_descent(initial_x, lr, iterations, gradient):
    x = initial_x
    trajectory = [x]
    for _ in range(iterations):
        grad = gradient(x)
        x -= lr * grad
        trajectory.append(x)
    return trajectory

```

```python
import matplotlib.pyplot as plt
import numpy as np

def plotGradientDescent(fArray, dfArray):
    x = np.linspace(-20,20,400)
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))
    for item in range(len(fArray)):
        trajectory =  batch_gradient_descent(-5, 0.01, 100, dfArray[item])
    #print(trajectory)
    
    
    
    # Plot for x^2
        axs[item].plot(x, f(x), label="Cost Function: x^2")
        axs[item].scatter(trajectory, fArray[item](np.array(trajectory)), c='red', label="Steps")
        for i in range(len(trajectory) - 1):
            axs[0].annotate('', xy=(trajectory[i + 1], f(trajectory[i + 1])),
                            xytext=(trajectory[i], f(trajectory[i])),
                            arrowprops=dict(arrowstyle='->', color='red', lw=1.5))
        axs[item].set_title("Batch Gradient Descent: x^2")
        axs[item].set_xlabel("x")
        axs[item].set_ylabel("Cost")
        axs[item].legend()

functionArray = [f,g]
dfArray =[ df,dg]
plotGradientDescent(functionArray,dfArray)
```
